outlier detection using IQR(Inter-Quartile Range)

```python
column = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
for col in column:
  Q1 = x[col].quantile(0.25) # first quartile
  Q2 = x[col].quantile(0.50) # second quartile
  Q3 = x[col].quantile(0.75) # third quartile
  IQR = Q3 - Q1 # inter-quartile range
  upper_bound = Q3 + 1.5* IQR # upper whisker
  lower_bound = Q1 - 1.5* IQR # lower whisker
  mask = (x[col]>=lower_bound) & (x[col]<=upper_bound) # datapoints that lie between upper whisker and lower whisker
x_outlier_detection = x[mask] # input features value after the removal of outlier
y_outlier_detection = y[mask] # target values after the removal of the outlier
```

outlier detection by the quantile method of the standard normal distribution

```python
Q = x_outlier_detection['Insulin'].quantile(0.95)
```

```python
Q
```

as the values under the insulin features have the maximum number of ouliers so we have to remove as much as the outliers only from the insulin feature so that our overall dataset will be modified out of the maximum outliers

```python
mask= x_outlier_detection['Insulin'] < Q
x_outlier_detection = x_outlier_detection[mask]
y_outlier_detection = y_outlier_detection[mask]
```

as we see the number of non diabetic (477) is greater than diabetic patients(244) so it will biase our model prediction more towards the non-diabetic which is not preferable for diagnosis.

** So we resampling the datapoints by SMOTE method where we put synthetic data to increase the minority datapoints**

** imbalanced learning**

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_smote, y_smote = smote.fit_resample(datanew,y_outlier_detection)
```

```python
x_smote.shape
```

```python
y_smote.shape
```

```python
y_smote.value_counts()
```

Training the model

splitting the dataset into the training dataset and testing dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size = 0.33, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

importing the logistic regression model and training the dataset

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

predicting on the test data of input features to evaluate the target value

```python
y_predict = model.predict(x_test)
```

```python
type(y_predict)
```

evaluation of the model prediction

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
accuracy = accuracy_score(y_test,y_predict)
print(accuracy)
confusion_matrix = confusion_matrix(y_test,y_predict)
print(confusion_matrix)
classification_report = classification_report(y_test,y_predict,target_names = ['Non-Diabetic','Diabetic'])
print(classification_report)
```
